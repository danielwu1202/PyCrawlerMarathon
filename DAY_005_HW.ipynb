```python
import requests
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text

response
```

## 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

## 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

## 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
import json
data = json.loads(response)

print('這個 API 一次會回傳',len(data),'筆資料'+'\n')
print('每一筆資料包含下列欄位:')
print(data[0].keys())
```

```python
#取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
for d in data:
    print('標題:', d['title'])
    print('貼文時間:', d['createdAt'])
    print('留言人數:', d['commentCount'])
    print('按讚人數:', d['likeCount'])
    print('')
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
comment_sum = 0
like_sum = 0

for d in data:
    comment_sum += d['commentCount']
    like_sum += d['likeCount']

print('熱門文章')
print('平均留言人數:', comment_sum / len(data))
print('平均按讚人數:', like_sum / len(data))
print('')

u = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
response_1 = u.text
data_1 = json.loads(response_1)

comment_sum_1 = 0
like_sum_1 = 0

for w in data_1:
    comment_sum_1 += w['commentCount']
    like_sum_1 += w['likeCount']

print('非熱門文章')
print('平均留言人數:', comment_sum_1 / len(data_1))
print('平均按讚人數:', like_sum_1 / len(data_1))
```
