# 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup
import re

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

# 如果想要依照類別分類，怎麼儲存會比較好？

```python
url = 'https://www.ettoday.net/news/news-list.htm'
html = requests.get(url)
soup = BeautifulSoup(html.text,'html.parser')

a = soup.find(class_="part_list_2").find_all('h3')
pattern = re.compile('tag\sc\w+')

#若依照分類，可以用字典的形式儲存。

news=dict()

for i in a:
    date = i.find(class_='date').text
    title = i.find('a').text
    tag = i.find(class_=pattern).text
    news.setdefault(tag,[])
    news[tag].append({date:title})
print(news)
```

```python
#上面跑出來的結果有'\u3000'為網頁原使碼，可用replace替換
news=dict()

for i in a:
    date = i.find(class_='date').text
    title = i.find('a').text
    #用''取代\u3000
    title = title.replace('\u3000','')
    tag = i.find(class_=pattern).text
    news.setdefault(tag,[])
    news[tag].append({date:title})
print(news)
```

# 哪一個類別的文章最多

```python
#新聞類型最多的是
total = dict()
for i in a:
    total[i.find(class_=pattern).text] = total.get(i.find(class_=pattern).text,0)+1
print(total)
M = max(total.values())
m = min(total.values())
for k,v in total.items():
    if total[k] == M:
        C = k
    if total[k] == m:
        c = k

print('最多類型的是{}，一共有{}則；最少類型的是{}，一共有{}則'.format(C,M,c,m))
```
